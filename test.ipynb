```python
import pandas
import numpy as np
import pandas as pd
data = pandas.read_csv('record_2019-01-01.csv')
data=np.array(data[0:500000])
print("shape date=",data.shape)
```

```python
new_data=pd.DataFrame()
for j in range (len(data)):
    #print("j",j)
    for i in range(j+1,len(data)):
        #print("i",i)
        if data[j,5]==data[i,5]:
            print("j=",j,"i=",i,data[j,4],data[i,4])
            new_data=new_data.append([data[j,:],data[i,:]])
            data=np.delete(data, i, 0)
            break
#print(len(data['userID'].unique()))
print(new_data.head(5))

for i in range (0,len(new_data), 2):
    for j in range(len(data)):
        if new_data.iloc[i,5]==data[j,5]:
            data=np.delete(data, j, 0)
            break
print(data.shape)
```

```python
for i in range (0,len(new_data), 2):
    for j in range(len(data)):
        if new_data.iloc[i,5]==data.iloc[j,5]:
            data = data.drop(data.index[j],axis=0)
            break
```
